# Creacion Consolidado por ROL

Comenzaremos el analisis de la data consolidando la info en un unico DataSet, para ello vamos a considerar los siguientes puntos:


- DATA:  Se considerará toda las causas que ingresaron al Sistema desde 2015 al 2019, esto ya que se cuenta además con la data de las audiencias en ese período.


- No se consideran causas anteriores a este período.

- Se considerarán Causas sin termino.

```python
import pandas as pd
import numpy as np
import os

from tqdm import tqdm
from unicodedata import normalize
```

```python
pd.set_option('display.max_columns', 100)
```

```python
path_processed = "../data/processed/pjud"
```

#### Paso 1 : Consolidacion de Dataset x rol

```python
df_ingresos_rol = pd.read_feather(f"{path_processed}/IngresosRol_feather")
print(f"{len(df_ingresos_rol)} causas en el dataset de ingresos por Rol")
```

```python
df_termino_rol = pd.read_feather(f"{path_processed}/TerminoRol_feather")
print(f"Existen : {len(df_termino_rol)} causas en el dataset de termino por rol")
```

Analizamos la data, uniendo ambos dataset.

```python
df_fulldata_rol = pd.merge(df_ingresos_rol, df_termino_rol, how='outer', on=['COD. TRIBUNAL','RIT'])
print(f"Existen: {len(df_fulldata_rol)} causas en total uniendo ambos dataset, aca existen datos sin registro de ingreso")
```

```python
df_causas_rol = pd.merge(df_ingresos_rol, df_termino_rol, how='inner', on=['COD. TRIBUNAL','RIT'])
print(f"Existen: {len(df_causas_rol)} causas ingresadas y terminadas")
```

```python
df_fulldata_rol.sort_values('FECHA INGRESO_x')
```

```python
df_fulldata_rol['MOTIVO TERMINO'].value_counts()
```

### Analisis de la Data Causas por ROL

De estos dos dataset debo considerar:

- Caso 1: el dataset df_causas_rol completo
- Caso 2: Debo agregar además las causas que tengan Fecha Ingreso_x pero que aún no terminan.
- Caso 3: Debo analizar las causas que no tienen Fecha Ingreso_x pero si tienen Fecha Ingreso_y y Fecha Ingreso_y, ya que deben ser las causas que no fueron tomadas por la consulta de Ingresos. Acá se debe reemplazar los datos obtenidos en Termino Materia a Ingresos.

Luego de tener estos tres dataset debo concatenarlos

### Caso 1:

```python
df_causas_rol
```

```python
df_causas_rol.columns
```

Eliminare columnas duplicadas para formar el dataset consolidado.

```python
columnas_drop = ['level_0_x', 'index_x', 'COD. CORTE_x', 'CORTE_x','TRIBUNAL_x', 'TIPO CAUSA_x', 'level_0_y','index_y']

df_causas_rol.drop(columnas_drop, axis = 'columns', inplace = True)
```

```python
df_causas_rol.rename(columns = {'COD. CORTE_y':'COD. CORTE',
                                'CORTE_y':'CORTE',
                                'TRIBUNAL_y':'TRIBUNAL',
                                'TIPO CAUSA_y':'TIPO CAUSA',
                                'FECHA INGRESO_x':'FECHA INGRESO IR',
                                'FECHA INGRESO_y':'FECHA INGRESO TR'
                                },  inplace = True)
```

```python
df_causas_rol.sort_values('FECHA INGRESO IR')
```

```python
df_causas_rol['MOTIVO TERMINO'].value_counts()
```

```python
causas_top = df_causas_rol[df_causas_rol['TRIBUNAL'].str.contains('ORAL')]
causas_top
```

```python
causas_top['MOTIVO TERMINO'].value_counts()
```

### Caso 2:

```python
filtro_sintermino = df_fulldata_rol['FECHA INGRESO_x'].notnull() & df_fulldata_rol['FECHA TERMINO'].isnull()
df_causas_sintermino = df_fulldata_rol.loc[filtro_sintermino]
```

```python
df_causas_sintermino
```

```python
df_causas_sintermino.columns
```

```python
columnas_drop = ['level_0_x', 'index_x','level_0_y', 'index_y', 'COD. CORTE_y', 'CORTE_y','TRIBUNAL_y', 'TIPO CAUSA_y']

df_causas_sintermino.drop(columnas_drop, axis='columns', inplace=True)
```

```python
df_causas_sintermino.rename(columns = {'COD. CORTE_x':'COD. CORTE',
                                        'CORTE_x':'CORTE',
                                        'TRIBUNAL_x':'TRIBUNAL',
                                        'TIPO CAUSA_x':'TIPO CAUSA',
                                        'FECHA INGRESO_x':'FECHA INGRESO IR',
                                        'FECHA INGRESO_y':'FECHA INGRESO TR'
                                       }, inplace = True)
```

```python
df_causas_sintermino
```

### Caso 3:

```python
filtro_sininicio = df_fulldata_rol['FECHA INGRESO_x'].isnull() & df_fulldata_rol['FECHA TERMINO'].notnull() 
df_causas_sininicio = df_fulldata_rol.loc[filtro_sininicio]
df_causas_sininicio.columns
```

```python
df_causas_sininicio
```

```python
columnas_drop = ['level_0_x', 'index_x', 'COD. CORTE_x', 'CORTE_x', 'TRIBUNAL_x', 'TIPO CAUSA_x', 'level_0_y', 'index_y']
df_causas_sininicio.drop(columnas_drop, axis='columns', inplace=True)
```

```python
df_causas_sininicio.rename(columns = {'COD. CORTE_y':'COD. CORTE',
                                        'CORTE_y':'CORTE',
                                        'TRIBUNAL_y':'TRIBUNAL',
                                        'TIPO CAUSA_y':'TIPO CAUSA',
                                        'FECHA INGRESO_x':'FECHA INGRESO IR',
                                        'FECHA INGRESO_y':'FECHA INGRESO TR'
                                       }, inplace = True)
```

```python
df_causas_sininicio
```

### Concatenar los dataset relacionados a ROL

```python
df_consolidado_rol = pd.concat([df_causas_rol,df_causas_sininicio,df_causas_sintermino], axis=0, ignore_index=True)
```

```python
df_consolidado_rol
```

ATENCION:
Se detecta que existen RIT 0-AÑO, lo cual no existe en la realidad, estos RIT deberían corresponder a errores del Sistema o ajustes para Metas de Gestión. Para efecto de este analisis deberían ELIMINARSE.

```python
df_rit_cero = df_consolidado_rol[df_consolidado_rol['RIT'].str.startswith("0-")]
```

```python
df_rit_cero
```

```python
df_consolidado_rol.drop(df_rit_cero.index, axis=0, inplace=True)
```

## CARGA DATA A ARCHIVOS FUTHER

```python
# Directorio donde se guardaran archivos feather
df_consolidado_rol.reset_index(inplace = True)
os.makedirs(path_processed, exist_ok = True)  

# Guardamos dataset como archivo feather

df_consolidado_rol.to_feather(f'{path_processed}/Consolidado_Rol_feather')
```

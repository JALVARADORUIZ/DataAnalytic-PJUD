# Creacion Consolidado por Materia Penal y Rol

Comenzaremos el analisis de la data consolidando la info en un unico DataSet

```python
import pandas as pd
import numpy as np
import os
import warnings

from tqdm import tqdm
from unicodedata import normalize
```

```python
pd.set_option('display.max_columns', 100)
```

```python
path_processed = "../data/processed/pjud"
```

```python
df_materia = pd.read_feather(f"{path_processed}/Consolidado_Materia_feather")
print(f"{len(df_materia)} causas en el dataset por materia")
```

```python
df_rol = pd.read_feather(f"{path_processed}/Consolidado_Rol_feather")
print(f"Existen : {len(df_rol)} causas en el dataset por Rol")
```

Analizamos la data, uniendo ambos dataset.

```python
df_union = pd.merge(df_materia, df_rol, how='outer', on=['COD. CORTE','COD. TRIBUNAL','RIT'])
print(f"Existen: {len(df_union)} causas en total uniendo ambos dataset")
```

```python
df_union.sort_values('FECHA INGRESO_x')
```

Existen Causas sin Materia ... 

```python
df_sin_materia = df_union[df_union['MATERIA'].isnull()]
df_sin_materia
```

Podemos ver que existen causas que estan con ROL pero no han sido incorporadas en causas por Materia.

¿Que hacer con ello?

A mi parecer se deben desestimar para el analisis, los datos son inherentes por su cantidad.

```python
df_sin_materia['MOTIVO TERMINO_y'].value_counts()
```

Estas no deberian tomarse en cuenta, ya que pueden deberse a causas mal tramitadas.

Veamos el caso especial donde no coinciden los MOTIVOS TERMINO

```python
df_diferencia = df_union[df_union['MOTIVO TERMINO_x']!= df_union['MOTIVO TERMINO_y']]
```

```python
df_diferencia['MOTIVO TERMINO_x'].value_counts()
```

```python
df_diferencia[df_diferencia['MOTIVO TERMINO_x'].isnull()]
```

```python
df_diferencia
```

Aca vemos el problema de :

que todos los delitos (materias) que no tienen término, obedecen a situaciones a las que el siagj no da respuesta, y dependiendo de los criterios de registro de la tramitación, dan origen a esta falta de término.

## CARGA DATA A ARCHIVOS FUTHER

```python
# Directorio donde se guardaran archivos feather
df_union.reset_index(inplace = True)
os.makedirs(path_processed, exist_ok = True)  

# Guardamos dataset como archivo feather

df_union.to_feather(f'{path_processed}/Consolidado_Materia_Rol_feather')
```

# Creacion Consolidado con audiencias

Comenzaremos el analisis de la data consolidando la info en un unico DataSet

```python
import pandas as pd
import numpy as np
import os
import warnings

from tqdm import tqdm
from unicodedata import normalize
```

```python
warnings.filterwarnings(action="ignore")
pd.set_option('display.max_columns', 100)
```

```python
path_processed = "../data/processed/pjud"
```

```python
df_causas = pd.read_feather(f"{path_processed}/Consolidado_Materia_Rol_feather")
print(f"{len(df_causas)} causas en el dataset por materia y rol")
```

```python
df_audiencias = pd.read_feather(f"{path_processed}/Audiencias_feather")
print(f"Existen : {len(df_audiencias)} audiencias")
```

```python
df_consolidado = pd.merge(df_causas,df_audiencias, how='outer', on=['COD. TRIBUNAL','RIT'])
```

```python
df_consolidado.columns
```

```python
df_audiencias[(df_audiencias['RIT']=='61-2019') & (df_audiencias['COD. TRIBUNAL']==991)]
```

## Carga Data relacionada a Poblacion

```python
df_poblacion = pd.read_feather(f"{path_processed}/DataConsolidada_Poblacion_Jurisdiccion_feather")
print(f"{len(df_poblacion)} registros")
```

```python
df_consolidado = pd.merge(df_consolidado,df_poblacion, how='outer', on=['TRIBUNAL'])
```

```python
df_consolidado.columns
```

```python
columnas_duplicadas = ['index_x', 'index_x','index_y','level_0','CORTE_x','index_y','index_x','index_y','CORTE_y']
df_consolidado.drop(columnas_duplicadas, axis='columns', inplace=True)
```

```python
df_consolidado
```

```python
df_consolidado[df_consolidado['TRIBUNAL_x']!=df_consolidado['TRIBUNAL_y']]
```

```python
# Directorio donde se guardaran archivos feather
df_consolidado.reset_index(inplace = True)
os.makedirs(path_processed, exist_ok = True)  

# Guardamos dataset como archivo feather

df_consolidado.to_feather(f'{path_processed}/Consolidado_FULL_feather')
```
